Key Fields in the Dataset
User Identifiers:

user_dim_nk: Unique user identifier.
email: Email address of the user.
hashed_email: Hashed version of the email for privacy.
Demographic Information:

enrich_firstname: First name of the user.
enrich_lastname: Last name of the user.
enrich_age: Age of the user (though in this case, it is missing).
enrich_gender: Gender of the user.
enrich_state: State where the user is located.
enrich_city: City where the user is located.
enrich_zip: ZIP code where the user is located.
enrich_address: Full address of the user.
country: Country of the user.
Engagement Data:

visit_count: Total number of visits by the user.
sports_with_visit_count: Number of sports the user has shown interest in, with the visit count for each.
top_sport: The sport with the highest engagement.
top_sport_visits: Number of visits related to the top sport.
entity_with_visit_count: Number of entities the user has interacted with, with visit counts.
top_entity: The entity with the highest engagement.
top_entity_visits: Number of visits related to the top entity.
latest_article_read: Most recent article read by the user.
total_engagement: Aggregate measure of user engagement.
touchpoint_count: Number of touchpoints with the user.
Content Interaction:

latest_sport_read: Most recent sport-related content read.
latest_entity_read: Most recent entity-related content read.
articles_read_last_30_days: Number of articles read by the user in the last 30 days.
Subscription Information:

newsletter_mail: Indicates if the user has a newsletter subscription.
no_of_newsletter_interested: Number of newsletters the user is interested in.
newsletter_counts: Total number of newsletters the user has interacted with.
Metadata:

first_login_date: Date when the user first logged in.
is_registered_user: Indicates if the user is registered.
last_updated: Date when the user's information was last updated
